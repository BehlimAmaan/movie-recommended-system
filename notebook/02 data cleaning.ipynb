```python
df = pd.read_csv(r"D:\Amaan_Behlim\03_Projects\Ml project\04_NLP\data\raw\movies_metadata.csv")
df.head()
```

```python
df.duplicated().sum()
df = df.drop_duplicates().reset_index(drop=True)
```

```python
df = df[['title','genres','overview','tagline','popularity','production_companies','vote_average']]
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['title'])
```

```python
df.isnull().sum()
```

```python
df['overview'] = df['overview'].fillna(" ") 
```

```python
df.head()
```

```python
import ast
df['genres'] = df['genres'].apply(lambda x:" ".join([i['name'] for i in ast.literal_eval(x)]))
df.head()
```

```python
df['tagline'] = df['tagline'].fillna(' ')
```

```python
df['production_companies'] = df['production_companies'].apply(lambda x:" ".join([i['name'] for i in ast.literal_eval(x)]))
```

```python
df.head()
```

```python
df['tags'] = df['overview'] + " " + df['genres'] + " " + df['tagline'] + " " + df['production_companies']
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words("english"))
lemit = WordNetLemmatizer()
```

```python
def processor(text):
    text = str(text).lower()
    text = re.sub(r'[^a-zA-Z\s]','',text)
    words = text.split()
    words = [word for word in words if word not in stop_words] 
    words = [lemit.lemmatize(word) for word in words]
    return " ".join(words)
# def preprocess_text(text):
#       #lowercase
#   text = str(text).lower()
#   #removed the punctuations
#   text = re.sub(r'[^a-zA-Z\s]', '', text)

#   words = text.split()
#   #remove stopwords
#   words = [word for word in words if word not in stop_words]

#   #lemmatize
#   words = [lemit.lemmatize(word) for word in words]

#   return " ".join(words)
```
