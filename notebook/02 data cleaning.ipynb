# 1. Handling Duplicated Values

```python
df.duplicated().sum()
df = df.drop_duplicates().reset_index(drop=True)
```

# 2. Feature Engineering  

```python
df = df[['title','genres','overview','tagline','popularity','production_companies','vote_average']]
df.head()
```

# 3. Handling Null Values

```python
df.isnull().sum()
```

```python
#title
df = df.dropna(subset=['title'])

#overview
df['overview'] = df['overview'].fillna(" ") 

#tagline
df['tagline'] = df['tagline'].fillna(' ')
```

# 4. Converting JSON-like columns


```python
import ast
#genres
df['genres'] = df['genres'].apply(lambda x:" ".join([i['name'] for i in ast.literal_eval(x)]))

#production_companies
df['production_companies'] = df['production_companies'].apply(lambda x:" ".join([i['name'] for i in ast.literal_eval(x)]))
df.head()
```

```python
#combining overview, genres, tagline and production_companies inot tags
df['tags'] = df['overview'] + " " + df['genres'] + " " + df['tagline'] + " " + df['production_companies']
```

# NLP
1. Lowercasing text
2. Removed the punctuations
3. Remove stopwords
4. Lemmatize
5. Vectorization


```python
import re, nltk
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words("english"))
lemit = WordNetLemmatizer()

def preprocess_text(text):
    #lowercase
    text = str(text).lower()

    #removed the punctuations
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    words = text.split()
    
    #remove stopwords
    words = [word for word in words if word not in stop_words]

    #lemmatize
    words = [lemit.lemmatize(word) for word in words]

    return " ".join(words)
```

```python
df['tags'] = df['tags'].apply(preprocess_text)
```

```python
df['tags'][0]
```

# Vectorization

```python
#TF-IDF

from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english'
                        , ngram_range=(1,2)
                        , max_features=50000
                        )

tfidf_matrix = tfidf.fit_transform(df['tags'])
```

```python
df = df.reset_index(drop = True)
indices = pd.Series(df.index, index = df['title']).drop_duplicates()
indices
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend(title, n=10):
    if title not in indices:
        return ['Movie Not Found']
    idx = indices[title]
    sim_score = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()
    similar_idx = sim_score.argsort()[::-1][1:n+1]
    return df['title'].iloc[similar_idx] 
```

```python
recommend('Toy Story')
```
